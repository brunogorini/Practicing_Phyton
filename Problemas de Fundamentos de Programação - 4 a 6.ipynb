# Problema 4

Faça um programa que processe matrizes de números de ponto flutuante com dimensões definidas pelo usuário e com valores gerados aleatoriamente num intervalo também definido pelo usuário. Escreva na saída padrão:

(A) A matriz gerada aleatoriamente;

(B) Qual é o maior valor contido na matriz e sua respectiva posição, linha e coluna. Casa haja empate, escreva um deles;
3

(C) Qual é o produto dos maiores valores de cada uma das linhas;

(D) Qual é a soma dos menores valores de cada uma das linhas;

(E) Qual a linha de maior intervalo entre seu menor valor e seu maior valor.

```python
# Subprogramas


def gerar():
    from random import uniform
    qtd_linhas = int(input('Quantidade de linhas: '))       # lê a quantidade de linhas
    qtd_colunas = int(input('Quantidade de colunas: '))     # lê a quantidade de colunas
    inicio = float(input('Inicio do intervalo: '))          # lê o início do intervalo de valores a serem gerados
    fim = float(input('Fim do intervalo: '))                # lê o fim do intervalo de valores a serem gerados

    mtx = [None] * qtd_linhas       # cria uma matriz vazia com a quantidade de linhas lida
    for i in range(qtd_linhas):
        mtx[i] = [0] * qtd_colunas  # adiciona a quantidade de colunas lida a matriz
        for j in range(qtd_colunas):
            mtx[i][j] = uniform(inicio, fim)    # adiciona os valores aleatórios gerados a matriz
    print('\n'+'A) A matriz gerada aleatoriamente:\n', mtx)          # imprime a matriz de números aleatórios gerada
    return mtx          # retorna esta matriz


def mostrar(mtx):
    max = (mtx[0][0], 0, 0)

    for i in range(len(mtx)):       # compara os valores contidos na matriz
        for j in range(len(mtx[i])):
            if mtx[i][j] > max[0]:
                max = (mtx[i][j], i, j)

    print('\n'+'B) O maior valor é %f' % max[0], 'na posicao (%d,' % max[1], '%d)' % max[2])     # imprime o maior valor


def prod_maiores(mtx):
    produto = 1
    for i in mtx:           # obtem os maiores valores de cada linha
        max = i[0]
        for j in i:
            if j > max:
                max = j
        produto *= max      # calcula o produto destes maiores valores

    print('\n'+'C) O produto dos maiores valores de cada uma das linhas:', produto)   # imprime este produto


def sum_menores(mtx):
    soma = 0
    for i in mtx:           # obtem os menores valores de cada linha
        min = i[0]
        for j in i:
            if j < min:
                min = j
        soma += min         # calcula a soma destes maiores valores

    print('\n'+'D) A soma dos menores valores de cada uma das linhas:', soma)    # imprime esta soma


def maior_linha(lin):
    maior = lin[0]
    for i in lin:
        if i > maior:
            maior = i   # obtem o maior valor de uma linha
    return maior        # retorna este valor


def menor_linha(lin):
    menor = lin[0]
    for i in lin:
        if i < menor:
            menor = i   # obtem o menor valor de uma linha
    return menor        # retorna este valor


def maior_interv(mtx):
    maior_intervalo = 0
    max_i = 0
    for i, linha in enumerate(mtx):
        interv = abs(menor_linha(linha) - maior_linha(linha)) # calcula o intervalo entre o menor e o maior valor de cada linha
        if interv > maior_intervalo:
            max_i = i                       # obtem a linha de maior intervalo
            maior_intervalo = interv        # obtem o valor do maior intervalo

    print('\n'+'E) A linha de maior intervalo entre seu menor valor e seu maior valor (linha %d):\n' % max_i, mtx[max_i])  # imprime


# Programa Principal

matriz = gerar()        # função que retorna uma matriz com dimensões definidas pelo usuário
mostrar(matriz)         # função que imprime o maior valor contido na matriz e sua respectiva posição
prod_maiores(matriz)    # função que imprime o produto dos maiores valores de cada uma das linhas
sum_menores(matriz)     # função que imprime a soma dos menores valores de cada uma das linhas
maior_interv(matriz)    # função que imprime a linha de maior intervalo entre seu menor valor e seu maior valor
```

# Problema 5

Faça um programa, contendo subprogramas, que leia pares de valores inteiros, x e y, até que o par de números zeros seja lido. Suponha que cada um destes pares representem um ponto no espaço bidimensional. Mantenha estes pontos como um vetor de tuplas (x,y). Após a entrada de todos pontos válidos, excetuando o par de zeros, escreva na saída padrão:

(A) A quantidade de pontos válidos;

(B) O vetor de pontos, escrevendo um ponto por linha;

(C) Caso existam, quais os dois pontos mais próximos entre si. Caso haja empate, escreva um deles;

(D) Caso existam, quais os dois pontos mais distantes entre si. Caso haja empate, escreva um deles;

(E) Caso existam, quais são as médias das coordenadas x e y.

```python
# Subprogramas


def le_pontos():
    lista = []  # cria um vetor vazio
    x, y = map(int, input('Digite dois inteiros separados por espaço: ').split())  # lê dois valores inteiros

    while (x != 0) and (y != 0):  # enquanto o par de números lido é diferente de zero
        lista.append((x, y))  # adiciona o par de números a lista
        x, y = map(int, input('Digite dois inteiros separados por espaço: ').split())  # lê próximo par de números

    print('\n'+'A) A quantidade de pontos válidos: %d' % len(lista))    # imprime a quantidade de pontos válidos
    return lista        # retorna vetor de pontos


def vetor_pontos(lista):        # imprime o vetor de pontos, escrevendo um ponto por linha
    print('\n'+'B) O vetor de pontos, escrevendo um ponto por linha:')
    if len(lista) == 0:
        print('[]')
    elif len(lista) == 1:
        print('[', str(lista[0][0]), ',', str(lista[0][1]), ']')
    elif len(lista) == 2:
        print('[', str(lista[0][0]), ',', str(lista[0][1]))
        print(' ', str(lista[len(lista) - 1][0]), ',', str(lista[len(lista) - 1][1]), ']')
    elif len(lista) >= 3:
        print('[', str(lista[0][0]), ',', str(lista[0][1]))
        for i in range(1, len(lista) - 1):
            print(' ', str(lista[i][0]), ',', str(lista[i][1]))
        print(' ', str(lista[len(lista) - 1][0]), ',', str(lista[len(lista) - 1][1]), ']')


def menor_dist(lista):
    menor_dist = ((lista[0][0] - lista[1][0]) ** 2 + (lista[0][1] - lista[1][1]) ** 2) ** 0.5
    pontos_prox = (lista[0][0], lista[0][1], lista[1][0], lista[1][1])
    for i in range(len(lista)-1):       # calcula a distância entre os pontos
        for j in range(i + 1, len(lista)):
            distancia = ((lista[i][0] - lista[j][0]) ** 2 + (lista[i][1] - lista[j][1]) ** 2) ** 0.5
            if distancia < menor_dist:
                menor_dist = distancia  # obtem a menor distância entre pontos
                pontos_prox = (lista[i][0], lista[i][1], lista[j][0], lista[j][1])  # obtem os dois pontos mais próximos
    print('\n'+'C) Os dois pontos mais próximos entre si são: (%d,' % pontos_prox[0], '%d)' % pontos_prox[1], 'e (%d,' % pontos_prox[2], '%d)' % pontos_prox[3])


def maior_dist(lista):
    maior_dist = ((lista[0][0] - lista[1][0]) ** 2 + (lista[0][1] - lista[1][1]) ** 2) ** 0.5
    pontos_dist = (lista[0][0], lista[0][1], lista[1][0], lista[1][1])
    for i in range(len(lista)-1):       # calcula a distância entre os pontos
        for j in range(i + 1, len(lista)):
            distancia = ((lista[i][0] - lista[j][0]) ** 2 + (lista[i][1] - lista[j][1]) ** 2) ** 0.5
            if distancia > maior_dist:
                maior_dist = distancia  # obtem a maior distância entre pontos
                pontos_dist = (lista[i][0], lista[i][1], lista[j][0], lista[j][1])  # obtem os dois pontos mais distantes
    print('\n'+'D) Os dois pontos mais distantes entre si são: (%d,' % pontos_dist[0], '%d)' % pontos_dist[1], 'e (%d,' % pontos_dist[2], '%d)' % pontos_dist[3])


def ponto_medio(lista):
    soma_x = 0
    soma_y = 0
    for i in range(len(lista)):
        soma_x += lista[i][0]   # soma os valores da coordenada x
        soma_y += lista[i][1]   # soma os valores da coordenada Y
    media_x = soma_x / len(lista)   # calcula a média dos valores da coordenada x
    media_y = soma_y / len(lista)   # calcula a média dos valores da coordenada y
    print('\n'+'E) O ponto médio é: (%4.2f,' % media_x, '%4.2f)' % media_y)   # imprime as médias das coordenadas x e y


# Programa Principal


list = le_pontos()  # função que lê os pontos inseridos e imprime a quantidade de pontos válidos
vetor_pontos(list)  # função que imprime o vetor de pontos

if len(list) > 1:
    menor_dist(list)    # função que imprime os dois pontos mais próximos entre si
    maior_dist(list)    # função que imprime os dois pontos mais distantes entre si
    ponto_medio(list)   # função que imprime as médias das coordenadas x e y (ponto médio)
```

# Problema 6

Faça um programa que leia pares de números inteiros, chamados decimal e base, do teclado até que um decimal negativo seja digitado ou uma base fora do intervalo 2 a 9 seja digitada. Com exceção do último par lido, que representa um par delimitador de fim, chame a função recursiva, chamada converte, que produza a string de conversão do decimal na nova base, ambos recebidos como parâmetro.

```python
# Subprograma


def converte(decimal, base):
    dig_invert = [decimal % base]   # cria uma lista de dígitos com o resto da divisão do decimal pela base
    while decimal // base != 0:     # enquanto divisão inteira do decimal pela base é diferente de zero
        decimal //= base            # calcula resto da divisão do decimal pela base
        dig_invert.append(decimal % base)   # adiciona resto da divisão a lista de dígitos
    dig_invert.reverse()        # inverte ordem de dígitos da lista
    resultado = ''              # cria string vazia
    for i in dig_invert:
        resultado += str(i)     # adiciona cada dígito como string a variável resultado
    return resultado            # retorna número convertido


# Programa Principal

num, bas = map(int, input('Digite dois inteiros separados por espaço:').split())    # lê dois números inteiros (decimal e base para conversão)

while (num >= 0) and (2 <= bas <= 9):   # enquanto decimal não é negativo e base está dentro do intervalo 2 a 9
    print(converte(num, bas))           # imprime número convertido pela função
    num, bas = map(int, input('Digite dois inteiros separados por espaço:').split())    # lê próximos dois números inteiros
```
